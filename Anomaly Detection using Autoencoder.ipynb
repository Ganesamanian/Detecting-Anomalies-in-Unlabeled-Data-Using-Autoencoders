# Feature Engineering

```python
# aggregate data by holding_party_id
business_data = transaction_df.groupby('holding_party_id').agg(
    total_transactions=('amount', 'count'),
    total_amount=('amount', 'sum'),
    avg_transaction_amount=('amount', 'mean'),
    std_transaction_amount=('amount', 'std'),
    unique_counterparties=('counter_party_id', 'nunique'),
    total_credits=('amount', lambda x: x[transaction_df['tx_type'] == 'credit'].sum()),
    total_debits=('amount', lambda x: x[transaction_df['tx_type'] == 'debit'].sum()),

).fillna(0).reset_index()

# Add a feature for credit-to-debit ratio
business_data['credit_debit_ratio'] = business_data['total_credits'] / (business_data['total_debits'] + 1e-5)  # Avoid division by zero

# Normalize the aggregated data
scaler = StandardScaler()
scaled_business_data = scaler.fit_transform(business_data.drop(columns=['holding_party_id']))
```

# Model Fitting

```python
# Build and train autoencoder for business-level anomalies
input_dim = scaled_business_data.shape[1]
encoding_dim = 8

input_layer = Input(shape=(input_dim,))
encoded = Dense(encoding_dim, activation='relu')(input_layer)
decoded = Dense(input_dim, activation='sigmoid')(encoded)

autoencoder = Model(inputs=input_layer, outputs=decoded)
autoencoder.compile(optimizer=Adam(learning_rate=0.001), loss='mse')

# Train the autoencoder
autoencoder.fit(scaled_business_data, scaled_business_data, epochs=50, batch_size=32, shuffle=True, validation_split=0.2)
plot_model(autoencoder, to_file='autoencoder_model_b.png', show_shapes=True)

# Predict and calculate reconstruction errors for businesses
business_reconstructions = autoencoder.predict(scaled_business_data)
business_mse = np.mean(np.square(scaled_business_data - business_reconstructions), axis=1)

# Set a threshold for business-level anomalies
business_threshold = np.percentile(business_mse, 95)  # Top 5% are anomalies
business_anomalies = business_mse > business_threshold

# Mark anomalous businesses
business_data['is_anomalous'] = business_anomalies

# Drill down into transactions of anomalous businesses
anomalous_business_ids = business_data[business_data['is_anomalous']]['holding_party_id']

print("Number of anomalous businesses detected:", len(anomalous_business_ids))
print("Anomalous businesses detected:", list(anomalous_business_ids))
```

Normal businesses exhibit a balanced or expected range of credit and debit transactions, while anomalous businesses deviate from this established pattern.
  
- **Low credit with high debit** suggests a potential liquidity issue or fraudulent activity, where outflows exceed inflows.

```python
# Filter transactions of anomalous businesses
anomalous_transactions = transaction_df[transaction_df['holding_party_id'].isin(anomalous_business_ids)]
print("Number of anomalous transaction suspected:", len(anomalous_transactions))
# Drop non-numeric columns
numeric_columns = anomalous_transactions.select_dtypes(include=['number']).columns

# Step 4: Detect transaction-level anomalies within anomalous businesses
# Normalize the transaction-level data for the anomalous businesses
anomalous_transactions_scaled = scaler.fit_transform(anomalous_transactions[numeric_columns])

# Define autoencoder architecture
input_dim = anomalous_transactions_scaled.shape[1]
encoding_dim = 8  # Adjust based on experimentation

input_layer = Input(shape=(input_dim,))
encoded = Dense(encoding_dim * 2, activation='relu', kernel_regularizer=l2(0.01))(input_layer)
# Adding Dropout
encoded = Dropout(0.2)(encoded)
encoded = Dense(encoding_dim, activation='relu', kernel_regularizer=l2(0.01))(encoded)  # Adding another layer
decoded = Dense(encoding_dim * 2, activation='relu', kernel_regularizer=l2(0.01))(encoded)
# Adding Dropout
encoded = Dropout(0.2)(encoded)
decoded = Dense(input_dim, activation='sigmoid')(decoded)

autoencoder = Model(inputs=input_layer, outputs=decoded)
autoencoder.compile(optimizer='adam', loss='mse')

# Train autoencoder
autoencoder.fit(anomalous_transactions_scaled, anomalous_transactions_scaled,
                epochs=50,  # Adjust epochs based on performance
                batch_size=128,
                shuffle=True,
                validation_split=0.2)

plot_model(autoencoder, to_file='autoencoder_model_t.png', show_shapes=True)
```

```python
# Reconstruct transactions using the trained autoencoder
reconstructed = autoencoder.predict(anomalous_transactions_scaled)
reconstruction_errors = np.mean(np.square(anomalous_transactions_scaled - reconstructed), axis=1)

# Set a threshold for anomalies (e.g., 95th percentile of errors)
transaction_threshold = np.percentile(reconstruction_errors, 95)
transaction_anomalies = reconstruction_errors > transaction_threshold
# Flag transactions as anomalies using .loc to avoid the SettingWithCopyWarning
anomalous_transactions['transaction_anomaly'] = transaction_anomalies
flagged_anomalous_transaction = anomalous_transactions[anomalous_transactions['transaction_anomaly'] == True]
print("Number of anomalous transaction detected:", len(flagged_anomalous_transaction))
# Add an 'anomaly' column to transaction_df
transaction_df['anomaly'] = transaction_df.index.isin(flagged_anomalous_transaction.index)
transaction_df.to_csv("transaction_with_anamoly_marked.csv")
```

The histogram above compares the distribution of normal and anomalous transactions, with both sets plotted on a log scale to better visualize the transaction amounts.

- **Normal Transactions (Blue):** These transactions form the bulk of the distribution. Most of them are concentrated in the lower transaction amount range, with a few extending into higher amounts. The log transformation helps to compress the scale and reveal the spread of lower-value transactions, which dominate the dataset.

- **Anomalous Transactions (Red):** Though anomalous transactions occur less frequently, they are spread across the entire range of transaction amounts. Some anomalies share similar transaction amounts as normal transactions, but others show extreme values, indicating that the anomalies could involve unusually high or low amounts. These transactions, even if infrequent, exhibit significant deviations from the typical distribution, which is often indicative of fraudulent or unusual behavior.

The log scale is crucial here as it allows us to observe patterns in both low-value and high-value transactions without being overwhelmed by the extremely large amounts, and it emphasizes how anomalous transactions, although fewer, span a wide range of values.
